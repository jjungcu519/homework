```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/20240909132716.csv', encoding='EUC-KR')
```

```python
df.head()
```

```python
age_groups = ['10대', '20대', '30대', '40대', '50대', '60대', '70세 이상']
time_columns = ['6:00', '6:20', '6:40', '7:00', '7:20', '7:40', '8:00', '8:20', '8:40', '9:00',
                '9:20', '9:40', '10:00', '10:20', '10:40', '11:00', '11:20', '11:40', '12:00', 
                '12:20', '12:40', '13:00', '13:20', '13:40', '14:00', '14:20', '14:40', '15:00', 
                '15:20', '15:40', '16:00', '16:20', '16:40', '17:00', '17:20', '17:40', '18:00', 
                '18:20', '18:40', '19:00', '19:20', '19:40', '20:00', '20:20', '20:40', '21:00', 
                '21:20', '21:40', '22:00', '22:20', '22:40', '23:00', '23:20', '23:40', '0:00', 
                '0:20', '0:40', '1:00', '1:20', '1:40', '2:00', '2:20', '2:40', '3:00', '3:20', 
                '3:40', '4:00', '4:20', '4:40', '5:00', '5:20', '5:40']
```

```python
# 필요한 열(시간대 및 연령대별 데이터)만 추출
filtered_df = df[df['연령별'].isin(age_groups)]
```

```python
# 연령대별 데이터를 시간대별로 가져오기
time_df = filtered_df[time_columns]

# 연령대별 데이터 추출
age_labels = filtered_df['연령별'].unique()
```

```python
plt.figure(figsize=(14, 8))
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

# 각 연령대의 시간대별 데이터를 선 그래프로 그리기
for idx, age_group in enumerate(age_groups):
    plt.plot(time_columns, time_df.iloc[idx, :], label=age_group)

# 그래프에 제목과 축 레이블 추가
plt.title('연령대별 시간대별 TV 이용율', fontsize=16)
plt.xlabel('시간대', fontsize=14)
plt.ylabel('TV 이용율 (%)', fontsize=14)
plt.xticks(rotation=90)  # 시간대 라벨이 겹치지 않도록 90도 회전
plt.legend(title='연령대')

# 그래프 표시
plt.tight_layout()
plt.show()
```
